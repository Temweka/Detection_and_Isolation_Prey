```python
import pandas as pd
import numpy as np
from os import listdir
import cv2
import matplotlib.pyplot as plt
```

```python
def load_data(directory):
    
    file_names = []
    
    # Iterate through each file in the directory
    for file in listdir(directory):
        # Check file name contains .wav
        if '.png' in file: #check extension
            # Append the file name
            file_names.append(file)
            
    return file_names
```

```python
def Filename_DataFrame(directory, label):
    list_of_files = load_data(directory)
    
    penguin_ID = []
    Video_ID = []
    Date = []
    Time = []
    Frame_No = []
    Label = []
    file_name = []

    for file in list_of_files:
        labels = file.split("_")
        penguin_ID.append(labels[0])
        Video_ID.append(labels[2][:-4])
        Date.append(labels[4])
        Time.append((labels[5]+ ":" + labels[6]+ ":" +labels[7]))
        Frame_No.append(labels[-1][:-4])
        Label.append(label)
        file_name.append(file)
        
    return pd.DataFrame({'File_Name':file_name, 'penguin_ID': penguin_ID,'Video_ID': Video_ID,'Date': Date, 'Time': Time,'Frame_No': Frame_No, 'Label': Label})
    
```
